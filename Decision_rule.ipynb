```python
import numpy as np
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
dataset = load_iris()
```

```python
dataset
```

```python
X = dataset["data"][:,[0,2]]
```

```python
X
```

```python
y = dataset["target"]
```

```python
len(X)
```

```python
y = np.where(y==0,0,1)
```

```python
def activation(x):
    if x >=1 :
        return 1
    else:
        return 0
```

```python
y
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
weight =  np.zeros((2,1))
bias = np.zeros((1,1))
```

```python
for _ in range(1000):
    for i in range(len(X)):
        output = np.dot(X[i],weight)+bias
        output = activation(output)
        error = output - y[i]
        weight -= 0.01 * error*np.array([X[i]]).T
        bias -= 0.01*error
```

```python
X_MIN,X_MAXX = X[:,0].min()-0.5, X[:,0].max()+0.5
```

```python
Y_MIN,Y_MAXX = X[:,1].min()-0.5, X[:,1].max()+0.5
```

```python
xx,yy = np.meshgrid(np.arange(X_MIN,X_MAXX,0.02),np.arange(Y_MIN,Y_MAXX,0.02))
```

```python
xx
```

```python
yy
```

```python
result = np.c_[xx.ravel(),yy.ravel()]
```

```python
result
```

```python
temp = []
for i in range(len(result)):
    temp.append(activation(weight[0][0]*result[i][0] + weight[1][0]*result[i][1] + bias))
```

```python
temp
```

```python
temp = np.array(temp)
```

```python
result = temp.reshape(xx.shape)
```

```python
result
```

```python
plt.contour(xx,yy,result)
plt.scatter(X[:,0],X[:,1],c = y)
```
